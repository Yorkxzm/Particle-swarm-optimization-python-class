```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

def fun1(x,y):
    z=np.sin(np.sqrt(x**2+y**2))/(np.sqrt(x**2+y**2))+np.exp((np.cos(2*np.pi*x)+np.cos(2*np.pi*y))/2)-2.71289
    return z
def fun2(x,y):
    A=10
    z=2*A+x**2-A*np.cos(2*np.pi*x)+y**2-A*np.cos(2*np.pi*y)
    return z
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
X,Y=np.meshgrid(x,y)
plt.rcParams['font.sans-serif']=['SimHei'] #
plt.rcParams['axes.unicode_minus']=False #
Z=fun1(X,Y)
fig=plt.figure()
ax=Axes3D(fig)
plt.rcParams['font.family'] = ['sans-serif']#
surf=ax.plot_surface(X,Y,Z,cmap=cm.coolwarm)
plt.show()
Z=fun2(X,Y)
fig=plt.figure()
ax=Axes3D(fig)
plt.rcParams['font.family'] = ['sans-serif']#
surf=ax.plot_surface(X,Y,Z,cmap=cm.coolwarm)
plt.show()
```

```python
import time
from psoclasses import basictestfun
from psoclasses.unconsoptipso import standardpso
#-1sovmax=True
upper=[10,10]
lower=[-10,-10]
dim=2
psize=100
iter_num=1000
def funs(X):
    x=X[:,0]
    y=X[:,1]
    z=-1*(np.sin(np.sqrt(x**2+y**2))/(np.sqrt(x**2+y**2))+np.exp((np.cos(2*np.pi*x)+np.cos(2*np.pi*y))/2)-2.71289)
    return z#max
adapso=standardpso(funs,lower,upper,dim,psize,ifplot=True,sovmax=True)
starttime=time.time()
adapso.fit()
endtime=time.time()
print(f"{endtime-starttime}s")
adapso.printresult()
```
